# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.   

**Описание данных:**  
airports_nearest — расстояние до ближайшего аэропорта в метрах (м)  
balcony — число балконов  
ceiling_height — высота потолков (м)  
cityCenters_nearest — расстояние до центра города (м)  
days_exposition — сколько дней было размещено объявление (от публикации до снятия)  
first_day_exposition — дата публикации  
floor — этаж  
floors_total — всего этажей в доме  
is_apartment — апартаменты (булев тип)    
kitchen_area — площадь кухни в квадратных метрах (м²)  
last_price — цена на момент снятия с публикации  
living_area — жилая площадь в квадратных метрах (м²)   
locality_name — название населённого пункта   
open_plan — свободная планировка (булев тип)   
parks_around3000 — число парков в радиусе 3 км  
parks_nearest — расстояние до ближайшего парка (м)   
ponds_around3000 — число водоёмов в радиусе 3 км  
ponds_nearest — расстояние до ближайшего водоёма (м)  
rooms — число комнат  
studio — квартира-студия (булев тип)  
total_area — площадь квартиры в квадратных метрах (м²)  
total_images — число фотографий квартиры в объявлении  
Пояснение: апартаменты — это нежилые помещения, которые не относятся к жилому фонду, но имеют необходимые условия для проживания.

### Обработка пропущенных значений

**Возможные причины пропусков значений и что будем делать с ними:**   
1. **ceiling_height** - высота потолков - люди, размещающие объявления могут не знать её, им не хочется измерять, считают что это не важный параметр.**Гипотеза:** высота потолков может коррелировать с местонахождением дома: центр СПб - старый фонд с высокими потолками, средний пояс - хрущёвки, окрайны - новостроки, и те и другие с низкими потолками.   
1. **floors_total**  - этажность дома - данные могут отсутствать по причине необязательности заполнения параметра. Учитывая, что  этажность здания могут коррелирует с местонахождением здания (центр СПб - старый фонд с высокими потолками, средний пояс - хрущёвки, окрайны - новостроки, и те и другие с низкими потолками), заполним эти данные в соответствии с этой логикой.  
1. **living_area** - жилая площадь - данные могут отсутствать по причине необязательности заполнения параметра.   
1. **kitchen_are**a - площадь кухни - данные могут отсутствать по причине необязательности заполнения параметра. **Гипотеза:** выясним, нет ли такой связи, что если квартира - студия, то там отсутствуют эти данные. Ведь в них нет явного деления площади на ту, которая относится к кухне, и ту, которая относится к жилой площади.   
1.  **is_apartment**  - апартаменты - люди могут не понимать разницы между квартирой и апартаметами. Данные могут отсутствать по причине необязательности заполнения параметра.  
1.  **balcony** - наличие балкона - вполне возможно, что где балкон отсутствует, там вместо 0 стоит NaN. Необходимо проверить, действительно ли по столбцу balcony присутствует только значение "1", говорящее, что балкон есть, а "0" - отсутствует.   
1.  **locality_name**  - населенный пункт - выглядит довольно странно, когда не указывают населенный пункт. Предположу, что свойственно это тем продавцам, которые уверены, что всем очевидно, что если они выставляют квартиру на продажу в разделе сайта "Санкт-Петербург", то и квартира в Санкт-Петербурге, не подозревая даже, что это может быть и область. ### День недели, месяц и год публикации объявления

```python
# Добавим колонку с годом - year публикации объявления
data['year'] = data['first_day_exposition'].apply(lambda x: x.year)
data['month'] = data['first_day_exposition'].apply(lambda x: x.month)
#display(data)
```

```python
# Добавим колонку с порядковым чилом месяцея - month публикации объявления
# Затем создадим словарь, чтобы изменить значение с чисел на название месяцев
# И поменяем значения на названия месяцев
data['month'] = data['first_day_exposition'].apply(lambda x: x.month)
months = {1: 'январь', 2: 'февраль', 3: 'март', 4: 'апрель', 
                    5: 'май', 6: 'июнь', 7: 'июль', 8: 'август', 9: 'сентябрь', 10: 'октябрь', 11: 'ноябрь', 12: 'декабрь'}
data['month'] = data['month'].map(months)
```

```python
# Добавим колонку с днём - weekday публикации
# Сначала проставим день недели в формате числа  от 1 - понедельник до 7 - воскресенье
# Затем создадим словарь, чтобы изменить значение с чисел на название дней недели
# И поменяем значения на название дней недели
data['weekday'] = data['first_day_exposition'].apply(lambda x: x.isoweekday())
weekdays = {1: 'понедельник', 2: 'вторник', 3: 'среда', 4: 'четверг', 
                    5: 'пятница', 6: 'суббота', 7: 'воскресенье'}
data['weekday'] = data['weekday'].map(weekdays)
#display(data)
```

### Удаление редких и выбивающиеся значений

```python
#data.describe() # Чтобы видеть как было
```

Выбросы - редкие и выбивающиеся значения, будем удалять на основе межквартильного метода.   
Хорошей статистикой для суммирования выборки распределений данных является метрика Interquartile Range (IQR). IQR рассчитывается как разница между 75-м и 25-м процентилями данных - IQR определяет средние 50% данных.    
Мы расчитаем 1 и 3 квартили  набора данных. Затем расчитаем IQR как разницу между 1 и 3 квартилями. Затем расчитаем пороговое значение для выбросов как 1,5-кратное IQR и вычтем это пороговое значение из 25-го процентиля и добавить его к 75-му процентилю, чтобы получить фактические пределы данных. Затем мы можем использовать эти пределы для определения значений выбросов и их удаления.

```python
def quartile_range(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq
#Столбцы с выбросами
list_features = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'days_exposition', 'ceiling_height', 'rooms']
for col in list_features:
    data[col] = quartile_range(data, col)

#Изучаем, есть ли выбивающиеся значения
data.describe()
```

### Анализ факторов влияющих на цену квартиры

#### Автоматизация процесса

Для автоматизации анализа напишем функцию, которая будет смотреть зависмость переменных и выводить в графическом виде 

```python
def impact_analysis(df, column):
    title_dict = {'total_area': 'Зависимость цены от площади','rooms': 'Зависимость цены от количества комнат', 'floor_level': 'Зависимость цены от этажа',
                  'category': 'Зависимость цены от удаленности от центра города',
                  'weekday':'Медиана цен по дням недели', 'month': 'Медиана цен по месяцам',
                  'year': 'Медиана цен по годам', 
                  'city_centers_km': 'Зависимость цены от удаленности от центра города'}
    plt.style.use('seaborn-pastel')
     
    if column == 'floor_level' or column == 'category':
        sort_key = 'last_price'
    else:
        sort_key = column
    if column == 'category':
        fontsize = 10
    else:
        fontsize = 10
    df_filtered = df.groupby(column).median()['last_price'].reset_index().sort_values(by=sort_key)
    plot = df_filtered.plot(x=column, y="last_price", kind='bar', ec='black', 
                            title=title_dict[column], legend=None, 
                            fontsize=fontsize, figsize=(20,4))
    ax2 = plt.gca()  # удалим xlabel - у нас и так есть многоговорящее название графика
    x_axis = ax2.xaxis
    if column != 'city_centers_km':
        plt.xticks(rotation=90)
 
    return plot
```

#### Зависимость цены от даты размещения

```python
def placement_price(df, ylims=None):
    dmy_to_plot = ['weekday', 'month', 'year']
    titles = ['Медиана цен по дню недели', 'Медиана цен по месяцу', 'Медиана цен по году']

    for i in range(len(dmy_to_plot)):
        df_to_plot = data.groupby(dmy_to_plot[i]).median()['last_price'].reset_index().sort_values(by=dmy_to_plot[i])
        if ylims is None:
            plot = df_to_plot.plot(x=dmy_to_plot[i], y="last_price", kind='bar', figsize=(10,5), ec='black',
                                   legend=False, title=titles[i], fontsize=15)
        else:
            plot = df_to_plot.plot(x=dmy_to_plot[i], y="last_price", kind='bar', figsize=(10,5), ec='black',
                               legend=False, title=titles[i], fontsize=15, ylim=ylims[i])
        ax2 = plt.gca()
        x_axis = ax2.xaxis
        x_axis.label.set_visible(False)
        if dmy_to_plot[i] != 'weekday':
            plt.xticks(rotation=90)
    return plot
```
